```python
import urllib2
from bs4 import BeautifulSoup
```

```python
url = 'http://esf.nanjing.fang.com/house-a0268-b02889/'
url = 'http://m.fang.com/esf/nanjing_268_b2889/'
```

```python
resp = urllib2.urlopen(url)
html = resp.read()
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
houseList = soup.findAll(attrs={"class":"houseList"})[0]
```

```python
houseList = houseList.findAll('li')
```

```python
for house in houseList:
    ps = house.findAll("p")
    print house.h3.text, ps[0].text, ps[1].text, "http:%s" % house.a['href']
```
